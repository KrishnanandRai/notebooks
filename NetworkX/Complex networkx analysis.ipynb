# 1. Visualization of networkx graphs

## 1.1 Using matplotlib

The simplest method is `nx.draw()`

```python
G = nx.dodecahedral_graph()
nx.draw(G)
```

For a lot more customization, use `nx.draw_networkx()`

```python
G = nx.dodecahedral_graph()
nx.draw_networkx(G, pos=nx.spring_layout(G))
# limits = plt.axis('off')
```

Have fun with `nx.draw_circular()` too.

```python
G = nx.complete_graph(25)
nx.draw_circular(G)
```

## 1.2 Using pygraphviz and pydot

PyGraphviz is a Python interface to the Graphviz graph layout and visualization package. With PyGraphviz you can create, edit, read, write, and draw graphs using Python to access the Graphviz graph data structure and layout algorithms. It is mainly maintained by NetworkX developers. Refer https://networkx.github.io/documentation/latest/reference/drawing.html for complete documentation

# 3. Algorithms

## 3.1 Isomorphism

We first generate a random Erdos-Renyi Graph with the 25 nodes and 0.2 as the probablity of edge creation.

```python
G = nx.fast_gnp_random_graph(n=25, p=0.2, seed=1) # Use '?' to look for the importance of seed.
```

```python
H = nx.fast_gnp_random_graph(n=25, p=0.2, seed=1)
```

```python
I = nx.fast_gnp_random_graph(n=25, p=0.2)
```

```python
nx.is_isomorphic(H, I)
```

```python
nx.is_isomorphic(G, I)
```

```python
nx.is_isomorphic(G, H)
```
